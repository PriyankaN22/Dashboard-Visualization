```python
import pandas as pd
import numpy as np
pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)
pd.set_option('max_rows', 300)
import pandas as pd
import numpy as nm
import matplotlib as mp
%matplotlib inline
import matplotlib.pyplot as plt
from ipywidgets import interact
import seaborn as sns
import pandas_datareader as pdr
from pandas_datareader.data import Options
import datetime as dt
import requests
import json
import urllib
```

```python
df7 = pd.read_excel("lab3_cleaned.xlsx")
```

Pass the entity name and get the ticker symbol for each entity

```python
def get_symbol(name):
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query={}&region=1&lang=en".format(name)
    
    result = requests.get(url).json()
    
    for x in result['ResultSet']['Result']:
        #if (x['exchDisp']=='NYSE'or x['exchDisp']=='NASDAQ'):
            #if x['typeDisp']=='Equity':
                    return x['symbol']

ticker=[]
cmp_name = df7.Entity

 
for xyz in cmp_name:
    ticker.append(get_symbol(xyz))
print(ticker)
```

Pass the ticker list to Pandas Reader to get the stock prices from Yahoo Finance

```python
start = dt.datetime(2004,1,1)
end = dt.datetime(2017,6,6)
data_source = 'yahoo'
panel_data = pdr.DataReader(ticker, data_source, start, end)
```

```python
panel_data
```

```python
close = panel_data.ix['Close']
all_weekdays = pd.date_range(start=start, end=end, freq='B')
close = close.reindex(all_weekdays)
close.head(10)
```

Convert the stock data into a csv file

```python
close.to_csv('Stocks.csv', index=True, encoding = 'utf-8')
```

```python
df_ticker = pd.DataFrame(ticker) 
df_ticker['Ticker_Symbol'] = pd.Series(ticker,index=df_ticker.index)
```

```python
df_ticker
```

Copy Ticker Symbol column into original data breaches dataframe

```python
df7['Ticker_symbol']= df_ticker['Ticker_Symbol']
```

```python
df7 = df7.dropna(axis=0,how='any')
```
