```python
import pandas as pd
import numpy as np
pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)
pd.set_option('max_rows', 300)



import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_excel('2017.xlsx')
```

```python
df
```

//Delete unnecessary columns

```python
df1=df.drop(df.columns[[1,2,7,10,11,12,13,14,15,16,17]], axis=1)
```

```python
df2 = df1.copy()
```

//Delete first row as it has unnecessary data

```python
df3= df2.iloc[1:]
```

//Records Lost column is not in number format so we will need to first
convert it to number format                                                    // ERROR1

```python
pd.to_numeric(df3['records lost'])
```

//Find out missing values in the file

```python
df3.isnull()
```

```python
df3.isnull().sum()
```

```python
df4 = df3.copy()
```


We need to address missing values in column "Records Lost". We see that there are two missing values. After analysis, it is found that "No of Records Stolen" is almost same as "Records Lost". These values differ for few records but the percentage difference is almost less than 10%. So we can fill missing values of "Records lost" with "No Of Records Stolen"

```python
df4['records lost'].fillna(df4['NO OF RECORDS STOLEN'],inplace=True)
```

//Once we have all the columns populated with what we need for our analysis, delete all the other unnecessary columns

```python
df5=df4.drop(df4.columns[5], axis=1)
```

//Convert year to an actual year

```python
df5.loc[df5.YEAR==0,'YEAR']=2004
df5.loc[df5.YEAR==1,'YEAR']=2005
df5.loc[df5.YEAR==2,'YEAR']=2006
df5.loc[df5.YEAR==3,'YEAR']=2007
df5.loc[df5.YEAR==4,'YEAR']=2008
df5.loc[df5.YEAR==5,'YEAR']=2009
df5.loc[df5.YEAR==6,'YEAR']=2010
df5.loc[df5.YEAR==7,'YEAR']=2011
df5.loc[df5.YEAR==8,'YEAR']=2012
df5.loc[df5.YEAR==9,'YEAR']=2013
df5.loc[df5.YEAR==10,'YEAR']=2014
df5.loc[df5.YEAR==11,'YEAR']=2015
df5.loc[df5.YEAR==12,'YEAR']=2016
df5.loc[df5.YEAR==13,'YEAR']=2017
df5.loc[df5.YEAR==14,'YEAR']=2017
```

//Map data sensitivity to an appropriate value

```python
df5.loc[(df5['DATA SENSITIVITY']==1),'DATA SENSITIVITY']='Just Email Addresses/Online Information'
df5.loc[(df5['DATA SENSITIVITY']==20),'DATA SENSITIVITY']='SSN/Personal Details'
df5.loc[(df5['DATA SENSITIVITY']==300),'DATA SENSITIVITY']='Credit card Information'
df5.loc[(df5['DATA SENSITIVITY']==4000),'DATA SENSITIVITY']='Email Passwords/Health Records'
df5.loc[(df5['DATA SENSITIVITY']==50000),'DATA SENSITIVITY']='Full Bank Account Details'
```

//Some of the records in "Data Sensitivity" column has incorrect value(typo error). For example, instead of 300 it is 3.
Also, some values instead of being in numeric format are in string format. We need to provide appropriate mapping for them   //ERROR2

```python
df5.loc[(df5['DATA SENSITIVITY']==3),'DATA SENSITIVITY']='Credit card Information'
```

```python
df5.loc[(df5['DATA SENSITIVITY']=='20'),'DATA SENSITIVITY']='SSN/Personal Details'
```

```python
df5.loc[(df5['DATA SENSITIVITY']=='300'),'DATA SENSITIVITY']='Credit card Information'
```

```python
df5
```

//Format heading

```python
df6 = df5.rename(columns={'YEAR':'Year','records lost':'Records Lost','ORGANISATION':'Organization','METHOD OF LEAK':'Method of Leak','DATA SENSITIVITY':'Data Sensitivity'})
```

//Insert a thousand separator for "Records Lost" column

```python
df6['Records Lost'] = df6.apply(lambda df6: "{:,}".format(df6['Records Lost']), axis=1)
```

```python
df6
```

```python
df6.to_csv("2017_clean_lab2.csv")
```
